Note that: even + even = even, uneven + even = uneven, uneven + uneven = even. 

Hence Fibonacci shows:
uneven, even, uneven, uneven, even, uneven, uneven, even, ...

We thus need to sum the values on places: $2+3*x$.

Explicit formula for Fibonacci on $n^{th}$ place is:

$$f_n = \frac{(1 + \sqrt{5})^n - (1 - \sqrt{5})^n}{2^n \sqrt{5}}$$
