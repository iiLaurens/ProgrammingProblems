#Project Euler
###Problem 31: Coin sums
In England the currency is made up of pound, £, and pence, p, and there are eight coins in general circulation:

1p, 2p, 5p, 10p, 20p, 50p, £1 (100p) and £2 (200p).
It is possible to make £2 in the following way:

1×£1 + 1×50p + 2×20p + 1×5p + 1×2p + 3×1p
How many different ways can £2 be made using any number of coins?
___

To make it easier we will consider cents, not pounds.

```python
%load_ext cython
```

```python
%%cython -a

import numpy as np
cimport numpy as np
import sys
cimport cython

cdef int[:] divisors = np.int_([1,2,5,10,20,50,100,200])

cdef int nr_coins = 8

@cython.boundscheck(False)
cdef int cdivide(int n,int m):
    cdef:
        int count = 0
        int i = 0
    
    while i < nr_coins and divisors[i] < n and divisors[i] <= m:
        count += cdivide(n-divisors[i],divisors[i])
        i += 1
            
    if i < nr_coins and divisors[i] == n and n <= m:
        count += 1
    
    return count

def divide(n):
    return cdivide(n,n)
```
